```python
# Importação da biblioteca Pandas
import pandas as pd
```

```python
# Leitura do arquivo CSV original
idh = pd.read_csv('./idh_2010_tratado.csv' , sep=';', encoding='latin-1')
```

```python
# Visualização do Dataframe original
idh
```

```python
# Exclusão das colunas que não serão utilizados no Dataframe
idh = idh.drop(columns=['IDHM'])
```

```python
# contagem total do número de resagintros do Dataframe 
idh.count()
```

```python
# Visualização do resultado final do pré processamento de dados
idh
```
